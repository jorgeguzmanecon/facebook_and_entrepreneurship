```python
import pandas  as pd
```

```python
df = pd.read_csv('/shared/share_scp/coresignal/coresignal_company.csv')
```

```python
url_clean = df["url"].astype("string").str.strip()
valid_url_mask = url_clean.notna() & url_clean.ne("")

duplicate_count = url_clean[valid_url_mask].duplicated().sum()
print("Duplicate url rows (excluding empty/null):", duplicate_count)
```
