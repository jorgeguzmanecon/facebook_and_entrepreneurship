```python
import pandas as pd
```

```python
df = pd.read_pickle("graduates_person_level_with_majors_latest.pkl")
```

```python
all_experience = pd.read_pickle('all_experience_AnalysisFile_latest.pkl')
all_experience.columns

```

```python

merged_data=  pd.merge(all_experience, df, left_on='member_id',right_on='linkedin_member_id', how='inner')
```

```python
merged_data.columns
```

```python
major_columns = [x for x in df.columns if 'major' in x]

print_columns = ['linkedin_member_id'] + major_columns

print_columns = ['title','company_name','company_url','linkedin_member_id', 'university_title','university_major_raw','is_founder_or_owner_with_url'] + major_columns

#df.sample(20)[[print_columns
pos = merged_data[merged_data['is_founder_or_owner_with_url']==1][print_columns].sample(10000)
neg = merged_data[merged_data['is_founder_or_owner_with_url']==0][print_columns].sample(10000)
pd.concat([pos,neg]).to_csv('./temp/founder_nonfounder_sample.csv', index=False)
```
